### <span style='color:darkmagenta'>About the Dataset(Superstore Sales Analysis)</span>

Here are the definitions for a few of the columns:
    
* **Ship Mode:** Mode of shipment used for shipment delivery.

* **Segment:** Customer segment where product was shipped to.

* **Country:** Country in which the shipment was delivered.

* **City:** City in which shipment was delivered.

* **State:** State in which shipment was delivered.

* **Postal Code:** Postal code the shipment was delivered to.

* **Region:** Country region.

* **Category:** The category product belongs to.

* **Sub-Category:** Sub-category of the product.

* **Sales:** Sales made.

### <span style='color:darkmagenta'>1. Customer Segmentation</span>

```python
# finding unique values from columns

# 1. Segment unique values
df['Segment'].unique()
```

```python
# 2. Segment value count

df['Segment'].value_counts()
```

- **Most Profitable Customer Segment** 

```python
Profitable_Cust_Seg= pd.DataFrame(df.groupby(['Segment'])[['Profit']].mean())
Profitable_Cust_Seg

# Most profitable customer segment is "Home Office"
```

```python
plt.figure(figsize=(10,6))
plt.rcParams.update({'font.size':14})
ax = sns.barplot(data = Profitable_Cust_Seg, x = Profitable_Cust_Seg.index, y = Profitable_Cust_Seg.Profit)
plt.title("Customer Segment Profitability",color='indigo',fontsize=12,fontweight='bold')
# Annotate the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 10), 
                textcoords='offset points')
plt.show()

'''Here is visualization of most profitable customer segment. From graph we can say that most profitable customer segment is "Home Office"'''
```

- **Most Profitable customer region:**

```python
profitable_region = pd.DataFrame(df.groupby(['Region'])['Profit'].sum().reset_index())
profitable_region
```

```python
plt.figure(figsize=(12,6))
plt.pie(profitable_region.Profit, labels = profitable_region.Region, startangle = 90, autopct = "%1.0f%%", explode = [0, 0, 0, 0.1])
plt.title("Most Profitable by Region", color='indigo',fontsize=14,fontweight='bold')
plt.show()

'''Most profitable Customer Segment is "West".'''
```

**Graph of count of Customer Segments**

```python
plt.figure(figsize=(15,8))
ax = sns.countplot(x=df.Segment)
plt.title('Count of Customer Segments',color='Black',fontsize=15,fontweight='bold')
plt.rcParams.update({'font.size':14});
# Annotate the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 10), 
                textcoords='offset points')
plt.show()

'''In Consumer segment cout is very high. But here we observe count of "Home Office" segment is very low. 
So improvement is need in this segment'''
```
